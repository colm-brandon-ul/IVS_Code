```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
edge_labels = nx.get_edge_attributes(G,"relationship") # key is edge, pls check for your case
formatted_edge_labels = {(elem[0],elem[1]):edge_labels[elem] for elem in edge_labels}
formatted_edge_labels
```

```python
pos = nx.kamada_kawai_layout(G)
nx.draw_networkx(G, with_labels=True, node_color="yellow", edge_labels=formatted_edge_labels)
#nx.draw_networkx_edge_labels(G,edge_labels=formatted_edge_labels,font_color='red')
plt.axis('off')
axis = plt.gca()
plt.rcParams['figure.figsize'] = [40, 40]
axis.set_xlim([1.2*x for x in axis.get_xlim()])
axis.set_ylim([1.2*y for y in axis.get_ylim()])
plt.savefig("graph.png")
plt.show()

```
